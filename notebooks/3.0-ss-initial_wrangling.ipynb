```python
# Plan: 
# pre-generate column names, in a way that can be scripted generated...
# for minutes in heartratezones each element in the array is a dictionary, so iterate through that, check the zoneName, then 
# for that column add the value. iterate in order of index of large array. Should end up with same length
```

```python
base_column_string = 'heartRateZones'
activity_df['heartRateZones.OutofRange.calories'] = None
activity_df['heartRateZones.FatBurn.calories'] = None
activity_df['heartRateZones.Cardio.calories'] = None
activity_df['heartRateZones.Peak.calories'] = None

activity_df['heartRateZones.OutofRange.minutes'] = None
activity_df['heartRateZones.FatBurn.minutes'] = None
activity_df['heartRateZones.Cardio.minutes'] = None
activity_df['heartRateZones.Peak.minutes'] = None


for index in range(0, len(activity_df['heartRateZones'])):
    data_list = activity_df['heartRateZones'].iloc[index]
    for small_dict in data_list:
        generated_base_col_string = base_column_string + '.' + small_dict['name']
        generated_base_col_string = generated_base_col_string.replace(" ", "")
        
        cal_val = small_dict['caloriesOut']
        #activity_df[generated_base_col_string + '.' + 'calories'].iloc[index] = cal_val
        activity_df.iloc[index, activity_df.columns.get_loc(generated_base_col_string + '.' + 'calories')] = cal_val
        
        minute_val = small_dict['minutes']
        #activity_df[generated_base_col_string + '.' + 'minutes'].iloc[index] = minute_val   
        activity_df.iloc[index, activity_df.columns.get_loc(generated_base_col_string + '.' + 'minutes')] = minute_val
    
```

# Skin Temperature Log

```python
# Take: dateTime, value
skin_df_arr = []
for i in range(1,10):
    file_path = f'../data/raw_data/get_skin_{i}.json'
    
    f = open(file_path)
    data = json.load(f)
    skin_df = pd.DataFrame.from_dict(pd.json_normalize(data['tempSkin']), orient='columns')
    skin_df = skin_df[['dateTime', 'value.nightlyRelative']]
    skin_df_arr.append(skin_df)
    
skin_df_arr[0].info()
```
