# Get Activity Log List

```python
get_activities_url = f'https://api.fitbit.com/1/user/-/activities/list.json'

get_activities_param = {
    #'beforeDate' : '2024-07-17',
    'afterDate' : '2024-07-18',
    'sort' : 'asc',
    'limit' : 100,
    'offset' : 0
}
get_activities_headers = {
    "authorization" : f"Bearer {os.getenv('ACCESS_TOKEN')}"
}

activities_req = requests.get(url = get_activities_url, params=get_activities_param, headers=get_activities_headers)

activities_json = activities_req.json()

print(activities_json)
```

```python
pprint(activities_json)
```

```python
# Active Zone Minutes are good for anyone wanting to meet guidelines from the World Health Organization, 
# the American Heart Association and various global bodies that recommend 150 minutes or 75 vigorous minutes 
# of exercise per week, explained Megan Ritter-Lluncor.

# To get an Active Zone Minute you need to elevate your heart rate into the fat-burn zone. 
# One minute gets you one minute towards your goal. However, if you get it higher than that, into Cardio 
# or Peak heart rate zones, you get double points. That means for every one minute of higher heart rate activity, 
# youll add two minutes towards your goal.

# Out of zone is below
# fat burn is resting heart rate plus 40% times the heart rate 
# Cardio is 60% and 
# peak is 85%

# https://www.wareable.com/fitbit/active-zone-minutes-explained-7968
```

```python
#activeDuration is the time in ms within each activity level (normal exercise plan, not doubled rate from elevated heart rate)
#activeZoneMinutes contains minutesInHeartRateZones which contains totalMinutes (double included), 
                                                                    # a list of the 4 types of zones, each has minutes, 
#activityLevel : sedentary. lightly, fairly, very... not sure wtf this means exactly, api has no definitions 
#activityName : sport or walk so far
#acctivityType : activityName id
#averageHeartRate : self explanatory
#calories
#duration: same as activeDuration if not manually edited
#hasActiveZoneMinutes : boolean, think this is always yes if not manually activity inputed.
```
